```python
import pandas as pd
import numpy as np

import json
import requests
from pandas.io.json import json_normalize
```

```python
#read webpage into pandas dataframe
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
print(df.shape)
df
```

```python
#create new dataframe without Borough values that equal "Not assigned"
df=df[df.Borough!='Not assigned']
df.head(11)
```

```python
df.shape
```

# Part 2 - adding Longitudes and Latitudes

```python
df_sorted=df.sort_values('Postal Code')
df_sorted
```

```python
locat=pd.read_csv('http://cocl.us/Geospatial_data')
print(locat.shape)
locat.head()
```

```python
#sorted version by Postal Code
toronto_zip=pd.merge(df_sorted,locat, how='outer', on='Postal Code')
toronto_zip.head(11)
```

```python
#unsorted version
toronto_zip2=pd.merge(df,locat, how='outer', on='Postal Code')
toronto_zip2.head(11)
```
